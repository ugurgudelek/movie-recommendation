```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as py
```

```python
df = pd.read_excel('results.xlsx')[:5]
```

```python
df
```

```python
df[['AE-SVD++','RAW-SVD++',
   'AE-SVD', 'RAW-SVD' ]].plot(figsize=(20,5), style={'AE-SVD++':'-.',
                                'RAW-SVD++':'-.',
                                'AE-SVD':':',
                                'RAW-SVD':':',
                                'AE-Mean':'-^',
                                'RAW-Mean':'-^',
                                'AE-KNNBasic':'-*',
                                'RAW-KNNBasic':'-*'})
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['AE-SVD++'],     '-.',    label=f'AE-SVD++ ($\mu:{df["AE-SVD++"].mean():.4f}$)', linewidth=4.0)
plt.plot(df['RAW-SVD++'],    '-.',    label=f'RAW-SVD++ ($\mu:{df["RAW-SVD++"].mean():.4f}$)')
plt.plot(df['AE-SVD'],       '-*',    label=f'AE-SVD ($\mu:{df["AE-SVD"].mean():.4f}$)', linewidth=4.0)
plt.plot(df['RAW-SVD'],      '-*',    label=f'RAW-SVD ($\mu:{df["RAW-SVD"].mean():.4f}$)')
plt.plot(df['AE-KNNBasic'],  '-^',    label=f'AE-KNNBasic ($\mu:{df["AE-KNNBasic"].mean():.4f}$)', linewidth=4.0)
plt.plot(df['RAW-KNNBasic'], '-^',    label=f'RAW-KNNBasic ($\mu:{df["RAW-KNNBasic"].mean():.4f}$)')
plt.xlabel('# of runs', fontsize=18)
plt.ylabel('RMSE', fontsize=24)
plt.legend()
plt.savefig('result.pdf')
plt.savefig('result.png')
```
